```python
# Step 0. Load libraries and custom functions
import duckdb
import pandas as pd
```

```python
# Step 1. Load the dataset
# 1.1 Read csv from webpage
df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

```python
# 1.2 Get the basic info
df.info()
```

```python
# 1.3 Sample 10 observations
df.sample(10)
```

```python
# 1.4 Let's use a SQL query to select the first 10 rows in the table
duckdb.query(
    "SELECT * FROM df LIMIT 10;"
)
```

```python
# 1.5 Show the results as a data frame
duckdb.query(
    "SELECT * FROM df LIMIT 10;"
).to_df()
```

```python
# 1.6 Retrieve the total number of deaths in Peru, until 31-12-2022
duckdb.query(
    "SELECT * FROM df WHERE location = 'Peru' LIMIT 10;"
).to_df()
```

```python
duckdb.query(
    "SELECT SUM(new_deaths) AS total_deaths FROM df WHERE location = 'Peru' OR location = 'Guatemala' AND date <= '2022-12-31' LIMIT 10;"
).to_df()
```
